```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
categorias = ['cat a', 'cat b','cat c', 'cat d']
valores =[25,40,30,90]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.bar(categorias, valores, zs=0, zdir='y', color='skyblue')

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')

ax.set_title('Grafico 3D')

plt.show()
```

```python
### COSTUMIZAO DO GRAFICO 3D

categorias = ['cat a', 'cat b','cat c', 'cat d']
valores =[25,40,30,90]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.bar(categorias, valores, zs=0, zdir='y', color=['skyblue','lightgreen','lightcoral','lightsalmon'],
       edgecolor='black')

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')

ax.set_title('Grafico 3D')

plt.show()
```

```python
### COSTUMIZAO DO GRAFICO 3D PART 2

categorias = ['cat a', 'cat b','cat c', 'cat d']
valores =[25,40,30,90]

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

ax.bar(categorias, valores, zs=0, zdir='y', color=['skyblue','lightgreen','lightcoral','lightsalmon'],
       edgecolor='black')
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.grid(False)

plt.xticks(rotation= 90, ha='right')

ax.view_init(elev=10,azim=15)

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')

ax.set_title('Grafico 3D')

plt.show()
```
