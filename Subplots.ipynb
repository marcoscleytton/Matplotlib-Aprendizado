```python
import matplotlib.pyplot as plt
import numpy as np

#Dados

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#Criação dos subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

#  Plot do primeiro subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Grafico 1')

#  Plot do primeiro subplot
axs[1].plot(x, y2, color='red')
axs[1].set_title('Grafico 2')

# Ajuste de layout
plt.tight_layout()

#Exibição
plt.show()
```
